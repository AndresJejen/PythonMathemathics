# Plotter Mapping complex vertical regions to sin(z)

This Code implements a plotter class that shows the mapping from complex vertical regions to sin(z), this code helps to understand how mapping works and let the user set the hyperparameters as:   
- **x_start** minimun x
- **x_end** maximun x
- **n_lines** How many vertical lines you want
- **legend** Let print the legend of the plot
- **y_start** y start (must be negative)
- **y_end** y end (must be positive)



```
import numpy as np
import matplotlib.pyplot as plt
```

```
class plot_sin_z():
  def __init__(self, x_start=-np.pi/2, x_end=np.pi/2, n_lines = 2, legend=False, y_start = -2, y_end= 2):
    self.x_start = x_start
    self.n_lines = n_lines
    self.x_end = x_end
    self.legend = legend
    self.y_start = y_start
    self.y_end = y_end
    self.valid = True
    self.__validate_params()

  def __validate_params(self):
    self.valid =  self.y_start <= 0 and self.y_end >= 0 and self.x_start <= 0 and self.x_end >= 0 and self.n_lines > 0
    if(self.valid == False):
      self.__print_params_error();

  def __print_params_error(self):
    print("ERROR - Verifica los parametros")

  def plot(self):
    if (self.valid):
      plt.figure()
      f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
      y = np.linspace(self.y_start,self.y_end,500)
      x_lines = np.linspace(self.x_start,self.x_end, self.n_lines)
      for i in x_lines:
        # plot part
        x = i*np.ones(500)
        ax1.plot(x,y, label = "Origin line $y_o = {0}$ , $y_f = {1}$, $x = {2}$".format(self.y_start, self.y_end, i))

        x = np.sin(x)*np.cosh(y)
        y_map = np.cos(x)*np.sinh(y)
        ax2.plot(x,y_map,label="mapped line from $y_o = {0}$ , $y_f = {1}$, $x = {2}$".format(self.y_start, self.y_end, i))
        # end plot part
      ax2.grid(True)
      ax1.axis('equal')
      ax2.axis('equal')
      ax1.grid(True)
      ax1.set_title(r'$z_0$ de origen')
      ax2.set_title(r'map de $sen(x_0)$')
      if (self.legend == True):
        ax1.legend()
        ax2.legend()
    else:
      self.__print_params_error()
```

```
graph = plot_sin_z(legend=True, n_lines=3)
graph.plot()
```
